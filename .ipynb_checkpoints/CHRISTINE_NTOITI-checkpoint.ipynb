##### 1.6. Data cleaning and PreprocessingCarry out dData cleaning and preprocessing steps below to ensureihe datat is ready for analysis.

Columns with more than 80% missing data (head_of_household_other columns with 98.69%, dreams_program column with 89.81%, 
dreams_program_othe column with  98.5%, and exit_reason_other column with 94.36%)  are a concern. 

All other columns will be imputed with appropriate strategies (median for numerical columns, and mode imputation for categorical columns).

```python
# Drop columns with more than 80% missing values
# Filter columns where missing data is greater than 80%
columns_to_drop = missing_data_percentage[missing_data_percentage > 80].index

# Drop those columns from the DataFrame
dreams_df = dreams_df.drop(columns=columns_to_drop)

# Fill missing values in categorical columns with 'Unknown' and numerical columns with mode
for column in dreams_df.columns:
    if dreams_df[column].dtype == 'object':  # Categorical column
        dreams_df[column] = dreams_df[column].fillna('Unknown')  # Fill ' '
    else:  # Numerical column
        dreams_df[column] = dreams_df[column].fillna(dreams_df[column].mode()[0])  # Fill with mode


# Check for any remaining missing values and handle
dreams_df.isnull().sum()


dreams_df.head(5)
```

#### 2.2. Visualizing the 'result' against other categories

```python
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_result_by_category(dreams_df, cat_variable, target_col='result'):
    """
    Visualizes the distribution of the 'result' column against a categorical variable.
    Annotates the bars with the count values for clarity.
    
    Parameters:
    dreams_df (DataFrame): The DataFrame containing the data.
    cat_variable (str): The categorical variable to compare 'result' against.
    target_col (str): The column representing the target variable, default is 'result'.
    
    Returns:
    None: Displays the plot.
    """
    # Strip any leading/trailing spaces in column names
    dreams_df.columns = dreams_df.columns.str.strip()
    
    # Check if the target column exists in the DataFrame
    if target_col not in dreams_df.columns:
        print(f"Warning: {target_col} does not exist in the DataFrame.")
        return

    # Visualize the distribution of 'result' by the categorical variable
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(x=cat_variable, hue=target_col, data=dreams_df, palette='Set2')
    
    # Add the count at the end of each bar
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width() / 2, height + 2,  # Positioning the text
                f'{int(height)}', ha='center', va='bottom', fontsize=12, color='black')  # Formatting text
    
    # Customize the plot
    plt.title(f'Distribution of {target_col} by {cat_variable}')
    plt.xlabel(cat_variable)
    plt.ylabel('Count')
    plt.xticks(rotation=90)
    
```

```python
# Visualize the result category against the county
visualize_result_by_category(dreams_df, 'county')
```

```python
# Visualize the result category against 'current_income_source'
visualize_result_by_category(dreams_df, 'current_income_source')
```

```python
# Visualize the result category against 'currently_in_school'
visualize_result_by_category(dreams_df, 'currently_in_school')
```

```python
# Visualize categorical columns
def visualize_categorical_distribution(dreams_df, cat_columns):
    """
    Visualizes the distribution of each categorical column in the DataFrame.
    """
    for col in cat_columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(y=col, data=dreams_df, palette='Set2')
        plt.title(f'Distribution of {col}')
        plt.xlabel('Count')
        plt.ylabel(col)
        plt.show()

categorical_columns = ['county', 'head_of_household', 'parents_alive_status', 'currently_in_school', 'ever_tested_for_hiv', 'result']
visualize_categorical_distribution(dreams_df, categorical_columns)
```
